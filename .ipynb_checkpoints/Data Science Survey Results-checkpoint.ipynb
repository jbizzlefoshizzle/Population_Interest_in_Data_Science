```python
# Import dependencies
import pandas as pd
import numpy as np

# Data to load
csv_data = "data/Topic_Survey_Assignment.csv"

# Define dataframe variable
survey_data = pd.read_csv(csv_data, index_col=0)
```

```python
# Show dataframe
survey_data
```

```python
# Sort data in descending order of Very interested
sorted_survey_data = survey_data.sort_values("Very interested", ascending=False)
sorted_survey_data
```

```python
# Convert numbers into percentages
total_respondents = 2233
temp_df = sorted_survey_data/total_respondents * 100

# Round percents to 2 decimal places
relative_frequency_df = temp_df.round(2)
relative_frequency_df
```

```python
# Matplotlib imports

import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot')
```

```python
relative_frequency_df.plot(kind='bar', # Bar graph
                           figsize=(20, 8), # using a figure size of (20, 8)
                           width=0.8, # bar width of 0.8
                           color=["#5cb85c", "#5bc0de", "#d9534f"])


# font size 14 for the bar labels, percentages, and legend
plt.xticks(size=14)
plt.yticks(size=14)

plt.legend(facecolor='w', # white background in legend
           fontsize=14)

plt.title("Percentage of Respondents' Interest in Data Science Areas", size=16) # font size 16 for the title

# hide gridlines
plt.grid(b=None)
# white background
background = plt.gca()
background.set_facecolor('w')

plt.show
```
